```python
import numpy as np

A = np.array([[5, 10, 15], [20, 25, 30]])  # Creating a 2D array using a nested list with values from 1–100

print("Shape:",A.shape)   # Using a property to get the structure of the array
print("max:",A.max())   # Using a method of the array to get the maximum value
print("max(A):",np.max(A)) # Using a NumPy function to get the maximum value
```

```python
B = np.ones((5, 5), dtype=int)  # Creates a 5x5 array filled with integer 1s

print(B)        # Displays the array
print("Shape:",B.shape)  # Shows the array has 5 rows and 5 columns
print("Size:", B.size)   # Shows the total number of elements (25)
print("NDim:", B.ndim)   # Shows the number of dimensions (2D array)
```

```python
C = np.arange(-5, 6)  # Creates values starting from -5 up to 5
C = C * 2             # Multiplies each value in the array by 2

print(C)              # Displays the final array
```
